```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
```

```python
#lendo o arquivo com o delimitador ;
df_2015 = pd.read_csv("Fluxo Escolar 2015.csv", encoding='latin1', delimiter=';')
```

```python
print(df_2015.columns)
```

```python
# Substituir vírgulas por pontos nas colunas numéricas e converter para float
colunas_numericas = ['CD_REDE_ENSINO','APR_1', 'REP_1', 'ABA_1', 'APR_2', 'REP_2', 'ABA_2', 'APR_3', 'REP_3', 'ABA_3']
for coluna in colunas_numericas:
    df_2015[coluna] = df_2015[coluna].astype(str).str.replace(',', '.').astype(float)
```

```python
# Agrupar e calcular a média das taxas de abandono por diretoria
abandono_por_diretoria_2015 = df_2015.groupby('NM_DIRETORIA')[['ABA_3']].mean()
```

```python
# Exibir os resultados
print(abandono_por_diretoria_2015)
```

```python
# Ordenar do maior para o menor
abandono_por_diretoria_2015 = abandono_por_diretoria_2015.sort_values(by='ABA_3', ascending=False)
```

```python
# Exibir os resultados
print(abandono_por_diretoria_2015)
```

```python
# Plotando as taxas de abandono por diretoria
abandono_por_diretoria_2015.plot(kind='bar', figsize=(14,8))
plt.title('Médias das Taxas de Abandono por Diretoria 2015')
plt.ylabel('Taxa de Abandono')
plt.xlabel('Diretoria')
plt.xticks(rotation=90)  # Rotacionar os nomes das diretorias
plt.legend(title='Abandono no ensino médio')
plt.tight_layout()
plt.show()
```
