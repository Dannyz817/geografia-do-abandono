```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
```

```python
anos = list(range(2013, 2024))
arquivos = [f"Fluxo Escolar {ano}_atualizado.csv" for ano in anos]
```

```python
dfs = []
```

```python
for arquivo in arquivos:
    df = pd.read_csv(arquivo)
    df['ANO_LETIVO'] = df['ANO_LETIVO'].astype(int)  # Converte a coluna para int
    dfs.append(df)
```

```python
# Concatena todos os DataFrames em um só
df_2013_2023 = pd.concat(dfs, ignore_index=True)
```

```python
# Salva o DataFrame consolidado em um arquivo CSV
df_2013_2023.to_csv("Fluxo Escolar 2013_2023.csv", index=False)
```

```python
# Faz o download do arquivo
files.download("Fluxo Escolar 2013_2023.csv")
```

```python
file_path = 'Fluxo Escolar 2013_2023.csv'
data = pd.read_csv(file_path)

# Calcular a média da coluna ABA_3 por ano
media = data.groupby('ANO_LETIVO')['ABA_3'].mean()

# Plotar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(media.index, media.values, marker='o', color='b', linestyle='-')
plt.title('Média de Abandono no Ensino Médio')
plt.xlabel('Ano')
plt.ylabel('Média de ABA_3')
plt.grid(True)
for year, value in media.items():
    plt.text(year, value, f'{value:.2f}', ha='center', va='bottom', fontsize=9)
plt.show()
```
