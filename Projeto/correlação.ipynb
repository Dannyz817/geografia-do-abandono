```python
# Importar bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
```

```python
# Carregar o arquivo CSV
df = pd.read_csv("Fluxo Escolar 2013_2023.csv")
```

```python
# Transformar os dados em um formato onde cada coluna é um ano de 2013 a 2023 para ABA_3
aba_3 = df.pivot_table(index=['NM_DIRETORIA', 'NM_MUNICIPIO'], columns='ANO_LETIVO', values='ABA_3')
```

```python
# Calcular a matriz de correlação
correlation_matrix_years = aba_3.corr()
```

```python
# Plotar o mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix_years, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlação de Abandono de 2013 a 2023")
plt.show()
```

```python
# Selecionar as colunas desejadas
selected_columns = df[['APR_3', 'REP_3', 'ABA_3']]

# Cálculo da correlação
correlation_matrix = selected_columns.corr()

# Exibindo a matriz de correlação no terminal
print("Matriz de Correlação:")
print(correlation_matrix)
```

```python
# Visualizando a matriz de correlação como heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de Correlação entre Aprovação, Reprovação e Abandono")
plt.show()
```
